Para este ejercicio, aportamos el código para inicializar el estado $|0_L\rangle$ del código de Steane:

```python
def Steane_init():
    # Definimos dos registros.
    # Registro con 7 qubits del código de Steane
    cq = QuantumRegister(7, 'q')

    # Registro con 6 qubit anzilla que guardará la
    # información sobre las medidas de los estabilizadores
    aq = QuantumRegister(6, 'a')

    # Registro clásico donde guardaremos los resultados
    # de medir los estabilizadores
    mb = ClassicalRegister(6, 'm')

    # Combinamos estos registros en un circuito
    qc = QuantumCircuit(cq, aq, mb)

    # Generamos el estado 0 lógico usando el circuito
    # dado en la hoja de problemas
    qc.h([0,1,5])
    qc.cx([0,5,1,0,5,1,0,1,5],[4,3,6,2,6,3,6,4,2])
    qc.barrier()
    return qc, cq, aq, mb
```

Podemos dibujar este circuito:

```python
qc, cq, aq, mb = Steane_init()
qc.draw('mpl', scale=0.5)
```

### 1. Define los subcircuitos encargados de obtener las medidas de los estabilizadores $S_z^{(1)}$, $S_z^{(2)}$ y $S_z^{(3)}$.

Las medidas de los estabilizadores $S_z^{(1)}$, $S_z^{(2)}$ y $S_z^{(3)}$ comparten una estructura similar entre ellas y con la medida de los estabilizadores que ya hemos visto en el Ejercicio 1. Por tanto, podemos definir una función general para estas medidas:

```python
def Sz_stabilizer(qc, code_qubits, anzilla_qubit):
    qc.cx(code_qubits, [anzilla_qubit])
    qc.barrier()
```

Podemos probar esta función para el estabilizador $S_z^{(1)}$:

```python
qc, cq, aq, mb = Steane_init()

# Introducimos el estabilizador Sz1
Sz_stabilizer(qc, [cq[0],cq[1],cq[2],cq[3]], aq[0])
qc.draw('mpl', scale=0.5)
```

### 2. Define los subcircuitos encargados de obtener las medidas de los estabilizadores $S_x^{(1)}$, $S_x^{(2)}$ y $S_x^{(3)}$.

Ya sabemos como realizar medidas de estabilizadores como los $S_z^{(1)}$, pero medir estabilizadores como $S_x^{(1)}$ puede ser menos intuitivo. Esto se puede conseguir utilizando puertas Hadamard para rotar la base $Z$ en la base $X$. Teniendo esto en cuenta se puede llegar a que un estabilizador de este tipo se puede medir usando una combinación de la medida de los estabilizadores $Z$ anteriores y puertas Hadamard:

```python
def Sx1_stabilizer(qc, code_qubits, anzilla_qubit):
    qc.h([0,1,2,3])
    qc.cx(code_qubits, [anzilla_qubit])
    qc.h([0,1,2,3])

cq = QuantumRegister(4, 'c')
aq = QuantumRegister(1, 'a')
qc = QuantumCircuit(cq, aq)
    
# Introducimos el estabilizador Sx1
Sx1_stabilizer(qc, [cq[0],cq[1],cq[2],cq[3]], aq[0])
qc.draw('mpl', scale=0.75)
```

Se puede simplificar el proceso de medida de estos estabilizadores reduciendo el número de puertas involucradas. Para esto se puede tener en cuenta que el circuito:

```python
qc = QuantumCircuit(2)
qc.h([0,1])
qc.cx(0, 1)
qc.h([0,1])

qc.draw('mpl')
```

Es equivalente al siguiente circuito:

```python
qc = QuantumCircuit(2)
qc.cx(1, 0)
qc.draw('mpl')
```

Teniendo en cuenta esto, obtén un circuito simplificado para la medida de los estabilizadores $S_x^{(1)}$, $S_x^{(2)}$ y $S_x^{(3)}$.

Teniendo en cuenta las pistas anteriores, se puede usar el circuito dado para medir los estabilizadores $X$ y modificarlo a base de quitar las puertas Hadamard a los lados de los qubits del código y añadirlas al principio y al final del circuito del qubit anzilla. También hay que cambiar el control de la puerta CNOT por el objetivo. Teniendo esto en cuenta, las medidas de los estabilizadores $S_x^{(1)}$, $S_x^{(2)}$ y $S_x^{(3)}$ comparten una estructura similar. Por tanto, podemos definir una función general para definir estas medidas:

```python
def Sx_stabilizer(qc, code_qubits, anzilla_qubit):
    qc.h(anzilla_qubit)
    qc.cx([anzilla_qubit], code_qubits)
    qc.h(anzilla_qubit)
    qc.barrier()
```

Podemos probar esta función para el estabilizador $S_x^{(1)}$:

```python
qc, cq, aq, mb = Steane_init()

# Introducimos el estabilizador Sx1
Sx_stabilizer(qc, [cq[0],cq[1],cq[2],cq[3]], aq[0])
qc.draw('mpl', scale=0.5)
```

Con estas funciones podemos definir el circuito que inicializa el estado $|0_L\rangle$ y realiza medidas de los estabilizadores:

```python
qc, cq, aq, mb = Steane_init()


# Introducimos el estabilizador Sz1
Sz_stabilizer(qc, [cq[0],cq[1],cq[2],cq[3]], aq[3])
# Introducimos el estabilizador Sz2
Sz_stabilizer(qc, [cq[1],cq[2],cq[4],cq[5]], aq[4])
# Introducimos el estabilizador Sz3
Sz_stabilizer(qc, [cq[2],cq[3],cq[5],cq[6]], aq[5])

# Introducimos el estabilizador Sx1
Sx_stabilizer(qc, [cq[0],cq[1],cq[2],cq[3]], aq[0])
# Introducimos el estabilizador Sx2
Sx_stabilizer(qc, [cq[1],cq[2],cq[4],cq[5]], aq[1])
# Introducimos el estabilizador Sx3
Sx_stabilizer(qc, [cq[2],cq[3],cq[5],cq[6]], aq[2])


anzillas = [aq[i] for i in range(6)]
bits = [mb[i] for i in range(6)]
qc.measure(anzillas, bits)

qc.draw('mpl', fold=70)
```

### 3. Define una función que, dados los resultados de las medidas de los estabilizadores, te devuelva que corrección hay que hacer.

Primero modificamos el código anterior para incluir la simulación y que nos devuelva el resultado de las medidas:

```python
qc, cq, aq, mb = Steane_init()


# Introducimos el estabilizador Sz1
Sz_stabilizer(qc, [cq[0],cq[1],cq[2],cq[3]], aq[3])
# Introducimos el estabilizador Sz2
Sz_stabilizer(qc, [cq[1],cq[2],cq[4],cq[5]], aq[4])
# Introducimos el estabilizador Sz3
Sz_stabilizer(qc, [cq[2],cq[3],cq[5],cq[6]], aq[5])

# Introducimos el estabilizador Sx1
Sx_stabilizer(qc, [cq[0],cq[1],cq[2],cq[3]], aq[0])
# Introducimos el estabilizador Sx2
Sx_stabilizer(qc, [cq[1],cq[2],cq[4],cq[5]], aq[1])
# Introducimos el estabilizador Sx3
Sx_stabilizer(qc, [cq[2],cq[3],cq[5],cq[6]], aq[2])


anzillas = [aq[i] for i in range(6)]
bits = [mb[i] for i in range(6)]
qc.measure(anzillas, bits)

# Simulamos medidas del circuito
backend = Aer.get_backend('qasm_simulator')
job = backend.run(qc, shots=1, memory=True)
stabilizer_measurements = job.result().get_memory()[0]
print("Medidas de los estabilizadores: {}".format(stabilizer_measurements))
qc.draw('mpl', fold=70)
```

Ahora podemos definir una función que tome el resultado de medir los estabilizadores y proponga una corrección:

```python
def Steane_correction(qc, stabilizer_measurements):
    # Los tres primeros bits de las medidas corresponden
    # a estabilizadores de tipo Z
    if stabilizer_measurements[:3] == '001': qc.x(0)
    if stabilizer_measurements[:3] == '010': qc.x(4)
    if stabilizer_measurements[:3] == '100': qc.x(6)
    if stabilizer_measurements[:3] == '011': qc.x(1)
    if stabilizer_measurements[:3] == '110': qc.x(5)
    if stabilizer_measurements[:3] == '101': qc.x(3)
    if stabilizer_measurements[:3] == '111': qc.x(2)
        
    # Los tres primeros bits de las medidas corresponden
    # a estabilizadores de tipo X
    if stabilizer_measurements[3:] == '001': qc.z(0)
    if stabilizer_measurements[3:] == '010': qc.z(4)
    if stabilizer_measurements[3:] == '100': qc.z(6)
    if stabilizer_measurements[3:] == '011': qc.z(1)
    if stabilizer_measurements[3:] == '110': qc.z(5)
    if stabilizer_measurements[3:] == '101': qc.z(3)
    if stabilizer_measurements[3:] == '111': qc.z(2)
```

Podemos añadir esta función al final de nuestro código anterior, lo cual completa la corrección de errores. Podemos comprobar que nuestro código es capaz de corregir hasta un error de tipo $X$ y uno de tipo $Z$ ocurriendo a la vez:

```python
qc, cq, aq, mb = Steane_init()

# Introducimos errores
qc.x(1)
qc.z(3)

qc.barrier()


# Introducimos el estabilizador Sz1
Sz_stabilizer(qc, [cq[0],cq[1],cq[2],cq[3]], aq[3])
# Introducimos el estabilizador Sz2
Sz_stabilizer(qc, [cq[1],cq[2],cq[4],cq[5]], aq[4])
# Introducimos el estabilizador Sz3
Sz_stabilizer(qc, [cq[2],cq[3],cq[5],cq[6]], aq[5])

# Introducimos el estabilizador Sx1
Sx_stabilizer(qc, [cq[0],cq[1],cq[2],cq[3]], aq[0])
# Introducimos el estabilizador Sx2
Sx_stabilizer(qc, [cq[1],cq[2],cq[4],cq[5]], aq[1])
# Introducimos el estabilizador Sx3
Sx_stabilizer(qc, [cq[2],cq[3],cq[5],cq[6]], aq[2])


anzillas = [aq[i] for i in range(6)]
bits = [mb[i] for i in range(6)]
qc.measure(anzillas, bits)
qc.barrier()

# Simulamos medidas del circuito
backend = Aer.get_backend('qasm_simulator')
job = backend.run(qc, shots=1, memory=True)
stabilizer_measurements = job.result().get_memory()[0]
print("Medidas de los estabilizadores: {}".format(stabilizer_measurements))

Steane_correction(qc, stabilizer_measurements)
qc.draw('mpl', fold=70)
```
